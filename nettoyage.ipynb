```python
# Import le csv
import csv
import pandas as pd
pd.set_option('display.max_columns', None)

path = "sante_mentale.csv"
df = pd.read_csv(path)
df
```

```python
df = df.replace('Oui / Yes', 'Oui').replace('Non / No', 'Non')
df
```

```python
df.columns
```

```python
# Supposant que `df` est votre DataFrame
df = df.iloc[20:]
df.info()
```

```python
df["Si oui, combien d'heures passes-tu à ce job par semaine ?"] = df["Si oui, combien d'heures passes-tu à ce job par semaine ?"].fillna(0)
df = df.dropna(subset=["Autour de quelle moyenne tourne tes notes à l'école ? Si tu en as"])
df["Si oui, bénéficies-tu d'un suivi médicale (psychiatre) pour ce diagnostic ?"] = df["Si oui, bénéficies-tu d'un suivi médicale (psychiatre) pour ce diagnostic ?"].fillna("Non")
df["Bénéficies-tu un suivi psychologique ?"] = df["Bénéficies-tu un suivi psychologique ?"].fillna("Non")
df = df.drop('Horodateur', axis=1)
```
