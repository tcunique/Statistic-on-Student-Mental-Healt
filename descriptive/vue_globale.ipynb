```python
import pandas as pd

# Charger les données depuis le fichier CSV
df = pd.read_csv('sante_mentale_clean.csv')

# Création d'une chaîne pour stocker les résultats
results = ""
for column in df.columns:
    unique_values = df[column].unique()  # Extraction des valeurs uniques
    num_unique_values = len(unique_values)  # Nombre de valeurs uniques
    results += f"Colonne '{column}':\n  Valeurs uniques: {unique_values}\n  Nombre d'entrées uniques: {num_unique_values}\n\n"

# Affichage des résultats dans le Notebook pour vérification
print(results)

# Spécifiez le chemin et le nom du fichier où enregistrer les résultats
filename = 'sante_mentale_clean_unique_values_results.txt'

# Écrire les résultats dans un fichier texte
with open(filename, 'w') as file:
    file.write(results)

print(f"Les résultats ont été sauvegardés dans le fichier {filename}")


```
