```python
import pandas as pd

# Charger les données depuis le fichier CSV
df = pd.read_csv('sante_mentale_clean.csv')

# Création d'une chaîne pour stocker les résultats
results = ""
for column in df.columns:
    unique_values = df[column].unique()  # Extraction des valeurs uniques
    num_unique_values = len(unique_values)  # Nombre de valeurs uniques
    results += f"Colonne '{column}':\n  Valeurs uniques: {unique_values}\n  Nombre d'entrées uniques: {num_unique_values}\n\n"

# Affichage des résultats dans le Notebook pour vérification
print(results)

# Spécifiez le chemin et le nom du fichier où enregistrer les résultats
filename = 'sante_mentale_clean_unique_values_results.txt'

# Écrire les résultats dans un fichier texte
with open(filename, 'w') as file:
    file.write(results)

print(f"Les résultats ont été sauvegardés dans le fichier {filename}")


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données depuis le fichier CSV
data = pd.read_csv('sante_mentale_clean.csv')

# Compter les occurrences de chaque âge
counts = data['age'].value_counts().sort_index()

# Extraire les âges et les nombres
ages = counts.index
counts = counts.values

# Créer le graphique en barres verticales
plt.figure(figsize=(10, 6))
bars = plt.bar(ages, counts, color='skyblue')
plt.xlabel('Âge')
plt.ylabel('Nombre de personnes')
plt.title('Répartition des âges')

# Ajouter les nombres totaux au-dessus de chaque barre
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{int(bar.get_height())}', 
             ha='center', va='bottom', color='black')

# Afficher le graphique
plt.tight_layout()  # Ajuster le layout pour éviter les coupures
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données depuis le fichier CSV
data = pd.read_csv('sante_mentale_clean.csv')

# Compter les occurrences de chaque ville d'étude
counts = data['ville_etude'].value_counts()

# Filtrer les villes d'étude avec un nombre de valeurs supérieur ou égal à 5
filtered_counts = counts[counts >= 5]

# Trier les villes d'étude par nombre de valeurs décroissant
sorted_counts = filtered_counts.sort_values(ascending=False)

# Extraire les villes d'étude et les nombres
villes = sorted_counts.index
counts = sorted_counts.values

# Créer le graphique en barres verticales
plt.figure(figsize=(10, 6))
bars = plt.bar(villes, counts, color='skyblue')
plt.xlabel('Ville d\'étude')
plt.ylabel('Nombre de personnes')
plt.title('Répartition des villes d\'étude (>= 5 occurrences)')

# Ajouter les nombres totaux au-dessus de chaque barre
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{int(bar.get_height())}', 
             ha='center', va='bottom', color='black')

# Afficher le graphique
plt.tight_layout()  # Ajuster le layout pour éviter les coupures
plt.show()
```
